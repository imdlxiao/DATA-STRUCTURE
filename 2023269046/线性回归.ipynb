编写一个计算闰年的代码
输入：2000
输出：Y

输入：1997
输出：N


```python
years = int(input("输入一个年份，判断是否是闰年"))
if years % 400 == 0 or (years % 100 != 0 and years % 4 ==0):
    print("Y")
else:
    print("N")
```

# 1、导入库文件

```python
import tensorflow as tf           #导入
import pandas     as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2、数据集导入

```python
data = pd.read_csv(r'E:\2023269046\Income.csv')
```

```python
data
```

```python
plt.scatter(data.Education, data.Income)
plt.xlabel('Education')
plt.ylabel("Income")
plt.title('2023269046')
```

# 3、模型搭建

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(1,) ))
```

```python
model.summary() # y=wx+b
```

# 4、训练配置

```python
model.compile(optimizer='adam',
              loss='mse',
              metrics='mse'
)
```

# 5、训练

```python
model.fit(data.Education,
         data.Income,
         epochs=10
)
```

```python
y_preed = model.predict(data.Education)

plt.scatter(data.Education,data.Income)
plt.plot(data.Education, y_preed, c='red')
plt.xlabel("Education")
plt.ylabel("Income")
plt.title("2023269046")
```

```python
model.fit(data.Education,
         data.Income,
         epochs=10000
)
```

```python
y_preed = model.predict(data.Education)

plt.scatter(data.Education,data.Income)
plt.plot(data.Education, y_preed, c='red')
plt.xlabel("Education")
plt.ylabel("Income")
plt.title("2023269046")
```

```python
model.fit(data.Education,
         data.Income,
         epochs=100
```

# 6、模型评估和预测

```python
model.evaluate(data.Education, data.Income)
```
